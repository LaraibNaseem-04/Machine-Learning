```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
df.head()
```

```python
df["target"] = iris.target
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
dfs= df.drop(columns = "target")
```

```python
X_train, X_test, y_train, y_test=train_test_split(dfs,df.target,test_size=0.25)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='rbf',C=30,gamma="auto")
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)  ##The problem is that the accuracy is changed again and again
```

```python
#now we try using loops
import numpy as np
```

```python
kernels = ["linear","rbf"]
c= [1,10,20]
average_scores = {}
for kval in kernels:
    for cval in c:
        cv_score = cross_val_score(SVC(kernel = kval,C=cval,gamma="auto"),iris.data,iris.target,cv=5)
        average_scores[kval+"    "+str(cval)]=np.mean(cv_score)
                       ##key                   #value
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(SVC(gamma="auto"),
                   {'C':[1,10,20],
                    'kernel':["rbf","linear"]},
                   cv=5,return_train_score=False) ##it does the same thing as cross validation but does all the manual work done in line50 manually
```

```python
clf.fit(iris.data,iris.target)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
model_params={
    "SVC":{
        "model":SVC(gamma="auto"),
        "params":{
            "kernel":["linear","rbf"],
            "C":[1,10,20]
        }
    },
    "randomforest":{
        "model":RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10]
        }
    },
    "logisticregression":
    {
        "model": LogisticRegression(solver="liblinear",multi_class = "auto"),
        "params":{
            "C":[1,5,10]
        }
    }
    }
```

```python
scores =[]
for model_name,mp in model_params.items():
    clf=GridSearchCV(mp["model"],mp["params"],cv=5,return_train_score=False)
    clf.fit(iris.data,iris.target)
    scores.append({
        "model":model_name,
        "best_score":clf.best_score_,
        "best_params":clf.best_params_
    })
```
