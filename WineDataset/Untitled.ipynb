```python
df = pd.DataFrame(wines.data , columns = wines.feature_names)
df
```

```python
df["target"] = wines.target
```

```python
targets = df.target
```

```python
df_inp = df.drop(columns = "target")
```

```python
df_inp
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df_inp,targets,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB
```

```python
modelG = GaussianNB()
```

```python
modelMt = MultinomialNB()
```

```python
modelG.fit(X_train,y_train)
```

```python
modelMt.fit(X_train,y_train)
```

```python
modelG.score(X_test,y_test)
```

```python
modelMt.score(X_test,y_test)
```

```python
##as we can see Gaussian has better performance here
```
